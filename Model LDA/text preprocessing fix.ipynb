## Tokenisasi

```python
# Tokenize
from nltk.tokenize import RegexpTokenizer
import nltk
nltk.download('punkt')
def word_tokenize_wrapper(text):
  tokenizer = RegexpTokenizer(r'\w+')
  return tokenizer.tokenize(text)

data_proses['token_berita']=data_proses['isi'].apply(word_tokenize_wrapper)
```

```python
print(data_proses['token_berita'][456])
```

## Remove Spesial karakter

```python
import re

def preprocess_text(tokens):
    # Menghapus karakter spesial dan tanda baca
    tokens = [re.sub(r'[^a-zA-Z]', '', token) for token in tokens]
    # Menghapus token kosong
    tokens = [token for token in tokens if token]
    # Menghapus token 1
    tokens = [token for token in tokens if len(token)>1]
    return tokens
data_proses['remove_punc'] =data_proses['token_berita'].apply(preprocess_text)
```

```python
print(data_proses['remove_punc'][456])
```

```python
list_token=[]
for token in data_proses['remove_punc']:
  list_token.extend(token)
len(list_token)
```

```python
cek_tokens = [token for tokens in data_proses['remove_punc'] for token in tokens if len(token) == 2]
word_counts = {}
for token in cek_tokens:
    if token in word_counts:
        word_counts[token] += 1
    else:
        word_counts[token] = 1

# Mengurutkan kata-kata berdasarkan frekuensi kemunculan dalam urutan menurun
sorted_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

# Menampilkan kata-kata dan frekuensinya
for word, count in sorted_words:
    print(f"Kata: {word}, Frekuensi: {count}")
```

## Stopword

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

list_stopwords =stopwords.words('indonesian')
txt_stopword =pd.read_csv("/content/drive/MyDrive/Bahan Penelitian/stopwords.txt", names=['stopwords'],header=None)
list_stopwords.extend(txt_stopword['stopwords'][0].split(' '))
list_stopwords=set(list_stopwords)

#remove stopword pada list token
def stopwords_removal(words):
    return [word for word in words if word not in list_stopwords]

data_proses['stopword_berita'] = data_proses['remove_punc'].apply(stopwords_removal)
```

```python
cek_tokens = [token for tokens in data_proses['stopword_berita'] for token in tokens if len(token) == 9]
word_counts = {}
for token in cek_tokens:
    if token in word_counts:
        word_counts[token] += 1
    else:
        word_counts[token] = 1

# Mengurutkan kata-kata berdasarkan frekuensi kemunculan dalam urutan menurun
sorted_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

# Menampilkan kata-kata dan frekuensinya
for word, count in sorted_words:
    print(f"Kata: {word}, Frekuensi: {count}")
```

## Bigram Model

```python
# Build the bigram models
import gensim
bigram = gensim.models.Phrases(data_proses['remove_punc'], min_count=50)
bigram_mod = gensim.models.phrases.Phraser(bigram)
```

```python
# Form Bigrams
data_proses['stopword_berita_bigram']= [bigram_mod[doc] for doc in data_proses['stopword_berita']]
```

```python
list_token=[]
for token in data_proses['stopword_berita_bigram']:
  list_token.extend(token)
len(list_token)
```

```python
cek_tokens = [token for tokens in data_proses['stopword_berita_bigram'] for token in tokens if len(token) == 20]
word_counts = {}
for token in cek_tokens:
    if token in word_counts:
        word_counts[token] += 1
    else:
        word_counts[token] = 1

# Mengurutkan kata-kata berdasarkan frekuensi kemunculan dalam urutan menurun
sorted_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

# Menampilkan kata-kata dan frekuensinya
for word, count in sorted_words:
    print(f"Kata: {word}, Frekuensi: {count}")
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
# Membuat objek stemmer
stemmer = StemmerFactory().create_stemmer()

# Daftar kata yang ingin dipertahankan
exceptions = ['pemilu','ketum']

# Fungsi untuk melakukan stemming pada kata-kata dalam list
def stem_list_of_words(word_list):
    stemmed_words = []
    for word in word_list:
        if word in exceptions:
            stemmed_words.append(word)
        else:
            stemmed_words.append(stemmer.stem(word))
    return stemmed_words

# Mengaplikasikan fungsi stem_list_of_words pada kolom 'stopword_berita'
data_proses['isi_stemmed'] = data_proses['stopword_berita_bigram'].apply(stem_list_of_words)
```

```python
print(data_proses['stopword_berita_bigram'][7283])
```

```python
print(data_proses['isi_stemmed'][7283])
```

```python
from itertools import chain
from collections import Counter
# Menggabungkan semua kata
isi_berita = data_proses['isi_stemmed']

# Menghitung frekuensi setiap kata
all_words = list(chain.from_iterable(isi_berita))
counter = Counter(all_words)

# Mengurutkan berdasarkan frekuensi kemunculan secara menurun
most_frequent = counter.most_common(30)

# Menampilkan 30 frekuensi kata tertinggi
for word, count in most_frequent:
    print(f"Kata: {word}\tFrekuensi: {count}")
```
