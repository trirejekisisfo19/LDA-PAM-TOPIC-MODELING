## **Olah Kompas**

**Buat Kolom Pada Masing-Masing Dataset**

```python
dataK1 = pd.read_csv("/content/drive/MyDrive/Bahan Penelitian/dataset penelitian/KOMPAS/scrape kedua/Anies.csv", header=None)
dataK1.columns=(['judul','tanggal','isi']) # menambahkan kolom baru sebagai header
```

```python
dataK2 = pd.read_csv("/content/drive/MyDrive/Bahan Penelitian/dataset penelitian/KOMPAS/scrape kedua/Ganjar.csv",header=None)
dataK2.columns=(['judul','tanggal','isi']) # menambahkan kolom baru sebagai header
```

```python
dataK3 = pd.read_csv("/content/drive/MyDrive/Bahan Penelitian/dataset penelitian/KOMPAS/scrape kedua/Prabowo.csv",header=None)
dataK3.columns=(['judul','tanggal','isi']) # menambahkan kolom baru sebagai header
```

```python
dataK4 = pd.read_csv("/content/drive/MyDrive/Bahan Penelitian/dataset penelitian/KOMPAS/scrape kedua/pilpres.csv",header=None)
dataK4.columns=(['judul','tanggal','isi']) # menambahkan kolom baru sebagai header
```

**Gabungkan File csv yang sudah ada Header**

```python
# Gabungkan file csv menggunakan concat
merged_df = pd.concat([dataK1, dataK2,dataK3,dataK4], axis=0,ignore_index=True)
```

**Ubah Format Tanggal**

```python
# Mengubah kolom tanggal menjadi tipe data datetime
merged_df['tanggal'] = merged_df['tanggal'].str.replace('Kompas.com -','').str.replace(', ','').str.replace(r'\d{1,2}\:\d{2} WIB', '', regex=True)
merged_df['tanggal'] = pd.to_datetime(merged_df['tanggal'], errors='coerce')
merged_df['tanggal'] = merged_df['tanggal'].dt.strftime('%d-%m-%Y')

# menghapus tanggal yg tidak sesuai format
indeks = [10,176,177,182,190,213]
merged_df = merged_df.drop(indeks)
merged_df = merged_df.reset_index(drop=True)

```

```python
# cek tipe data
merged_df['tanggal'] = pd.to_datetime(merged_df['tanggal'], errors='coerce')
merged_df.dtypes
```

**Hapus Tanggal yg Tidak diPerlukan**

```python
# membuat rentang pengambilan data
batas_awal = pd.to_datetime('2022-10-1')
batas_akhir = pd.to_datetime('2023-03-31')
data_pakai = merged_df.loc[(merged_df['tanggal'] >= batas_awal) & (merged_df['tanggal'] <= batas_akhir)]
data_pakai = data_pakai.reset_index(drop=True)
data_pakai
```

**Export File**

```python
data_pakai.to_csv('data_Kompas.csv',index=False)
```

## **Olah Tribuns**

**Buat Kolom Pada Masing" Dataset**

```python
dataT1 = pd.read_csv("/content/drive/MyDrive/Bahan Penelitian/dataset penelitian/TRIBUNNEWS/scrap kedua/Anies.csv", header=None)
dataT1.columns=(['judul','tanggal','isi']) # menambahkan kolom baru sebagai header
```

```python
dataT2 = pd.read_csv("/content/drive/MyDrive/Bahan Penelitian/dataset penelitian/TRIBUNNEWS/scrap kedua/Ganjar.csv", header=None)
dataT2.columns=(['judul','tanggal','isi']) # menambahkan kolom baru sebagai header
```

```python
dataT3 = pd.read_csv("/content/drive/MyDrive/Bahan Penelitian/dataset penelitian/TRIBUNNEWS/scrap kedua/Prabowo.csv", header=None)
dataT3.columns=(['judul','tanggal','isi']) # menambahkan kolom baru sebagai header
```

```python
dataT4 = pd.read_csv("/content/drive/MyDrive/Bahan Penelitian/dataset penelitian/TRIBUNNEWS/scrap kedua/pilpres.csv", header=None)
dataT4.columns=(['judul','tanggal','isi']) # menambahkan kolom baru sebagai header
```

**Gabungkan File csv yg sudah ada header**

```python
# Gabungkan file csv menggunakan concat
tribun_df = pd.concat([dataT1, dataT2,dataT3,dataT4], axis=0,ignore_index=True)
tribun_df
```

**Ubah Format Tanggal**

```python
# Mengubah kolom tanggal menjadi tipe data datetime
import dateparser

tribun_df['tanggal'] = tribun_df['tanggal'].str.replace(' WIB','').str.replace(r'^\w+, ', '', regex=True).str.replace(r' \d{1,2}\:\d{2}', '', regex=True)
tribun_df['tanggal'] = tribun_df['tanggal'].apply(lambda x: dateparser.parse(x).strftime('%Y-%m-%d') if x and dateparser.parse(x) else '')
tribun_df['tanggal'] = pd.to_datetime(tribun_df['tanggal'], format='%Y-%m-%d')

```

```python
# cek tipe data
tribun_df.dtypes
```

**Hapus Tanggal yang tidak diperlukan**

```python
# membuat rentang pengambilan data
batas_awal = pd.to_datetime('2022-10-1')
batas_akhir = pd.to_datetime('2023-03-31')
data_tribun = tribun_df.loc[(tribun_df['tanggal'] >= batas_awal) & (tribun_df['tanggal'] <= batas_akhir)]
data_tribun = data_tribun.reset_index(drop=True)
data_tribun
```

```python
# export ke file csv baru
data_tribun.to_csv('data_Tribun.csv',index=False)
```
